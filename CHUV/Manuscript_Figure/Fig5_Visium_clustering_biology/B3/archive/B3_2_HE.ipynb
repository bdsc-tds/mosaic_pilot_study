```python
import img2pdf
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
tif_path = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Data/Visium/High_Res_HE/B3-2.tif"
tif_crop_path = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Results/Manuscript_Figures_Final/Fig5/B3/archive/B3_2_HE.tif"
tif_crop_pdf = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Results/Manuscript_Figures_Final/Fig5/B3/B3_2_HE.pdf"
```

```python
# img = Image.open(tif_path) # exceeds limit
img=mpimg.imread(tif_path)
imgplot = plt.imshow(img)
```

```python
# Open and crop the image
with Image.open(tif_path) as img:
    left, upper, right, lower = 3000, 800, 9000, 8750
    cropped_img = img.crop((left, upper, right, lower))
plt.imshow(cropped_img)
```

```python
cropped_img.save(tif_crop_path)
```

```python
# Convert to PDF using img2pdf
with open(tif_crop_pdf, "wb") as f:
    f.write(img2pdf.convert(tif_crop_path))
```

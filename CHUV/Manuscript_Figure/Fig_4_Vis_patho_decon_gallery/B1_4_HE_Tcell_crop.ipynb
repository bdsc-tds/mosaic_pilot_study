```python
import img2pdf
from PIL import Image
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
tif_path = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Data/Visium/High_Res_HE/B1-4.tif"
tif_crop_path = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Results/Manuscript_Figures_Final/Fig4/B1_cropped/archive/B1_4_HE_Tcell.tif"
tif_crop_pdf = "/work/PRTNR/CHUV/DIR/rgottar1/spatial/Owkin_Pilot_Results/Manuscript_Figures_Final/Fig4/B1_cropped/B1_4_HE_Tcell.pdf"
```

```python
# img = Image.open(tif_path) # exceeds limit
img=mpimg.imread(tif_path)
imgplot = plt.imshow(img)
```

```python
# Open and crop the image
with Image.open(tif_path) as img:
    left, upper, right, lower = 5515, 850, 6855, 1935
    cropped_img = img.crop((left, upper, right, lower))
plt.imshow(cropped_img)
```

```python
cropped_img.save(tif_crop_path)
```

```python
# Convert to PDF using img2pdf
with open(tif_crop_pdf, "wb") as f:
    f.write(img2pdf.convert(tif_crop_path))
```

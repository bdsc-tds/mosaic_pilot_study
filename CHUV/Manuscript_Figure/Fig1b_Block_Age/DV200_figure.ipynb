```python
import pandas as pd
from plotnine import *
import plotnine as p9
```

```python
qc = pd.read_csv("metadata_blocage_dv200.tsv", sep="\t")
```

```python
qc
```

```python
p9.options.figure_size = (6.5, 4)

plot = (ggplot(qc) + geom_point(aes(x="Block_age_month", y="DV200", fill="Histology", shape="Sampling_Type"), size=5) +
        theme_bw() +
        guides(fill=guide_legend(ncol=1)) +
        scale_fill_manual(values={"DLBCL":"#FFD700", "Lung squamous carcinoma":"#FFA44F", "Lung adeno-carcinoma":"#B37337",
                                  "Breast lobular carcinoma":"#00A674", "Breast mucinous carcinoma":"#004D36", "Breast ductal carcinoma":"#00FFB3"}) +
        theme(legend_position='bottom', legend_direction="horizontal") +
        guides(fill=guide_legend(nrow=2, order=1), shape=False) +
        scale_shape_manual(values={"Resection": "o", "Biopsy": "^"}) +
        theme(axis_text=element_text(size=12)) +
        theme(plot_title=element_text(size=14))
)
```
